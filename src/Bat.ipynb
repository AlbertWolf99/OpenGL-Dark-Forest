```python
#Keyboard and Mouse Events Capture
boolScale = 0
s_x = 1.0
s_y = 1.0
t_x = 0.0
t_y = 0.0
theta = 0
def key_event(window,key,scancode,action,mods):
    global t_x, t_y
    global s_x, s_y
    global boolScale
#     print('[key event] key=',key)
#     print('[key event] scancode=',scancode)
#     print('[key event] action=',action)
#     print('[key event] mods=',mods)
#     print('-------')
    if key == 262: #right
        t_x += 0.01
    elif key == 263: #left
        t_x -= 0.01
    if key == 264: #down
        t_y -= 0.01
    elif key == 265: #up
        t_y += 0.01
    if key == 68: # D
        s_x += 0.01
    elif key == 65: # A
        s_x -= 0.01
    if key == 83: # S
        s_y -= 0.01
    elif key == 87: # W
        s_y += 0.01
        
    elif key == 77: # M
        boolScale = True
    

def mouse_event(window,button,action,mods):
    global theta
    #print('[mouse event] button=',button)
    #print('[mouse event] action=',action)
    #print('[mouse event] mods=',mods)
    #print('-------')
    if button == 0:
        theta -= 0.01
    elif button == 1:
        theta += 0.01
```

```python
def getCircleCoordinates(n_vertices, radius, center = (0, 0)):
    '''
    Get coordinates to draw a circle.
    Parameters
    ----------
        n_vertices : int
            Number of vertices that will be used to draw the circle.
        radius : float
            Radius of circle
        center : (float, float)
            Center of circle
    Return
    ------
        vertices
            Vertices with coordinates to draw the circle
    '''
    
    pi = 3.14
    angle = 0.0
    vertices = []
    
    for counter in range(n_vertices):
        angle += 2*pi/n_vertices
        x = center[0] + math.cos(angle) * radius
        y = center[1] + math.sin(angle) * radius
        vertices.append((x,y))
    
    return vertices
```

```python
def setColor(loc_color, color):
    glUniform4f(loc_color, color[0], color[1], color[2], color[3])
```

```python
def multiply_matrix(a,b):
    m_a = a.reshape(4,4)
    m_b = b.reshape(4,4)
    m_c = np.dot(m_a,m_b)
    c = m_c.reshape(1,16)
    return c
```
